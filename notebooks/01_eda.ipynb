## Proyecto Prediccion de atraso de vuelos ##

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Carga de datos
df = pd.read_csv("flight_data.csv")
print(df.shape)
print(df.head())

# 2. Exploración inicial
sns.histplot(df['dep_delay'], bins=50, kde=True)
plt.title("Distribución de retrasos en salida")
plt.show()

# 3. Limpieza de datos
# En 01_eda.ipynb
df_clean = df.drop(['arr_time','sched_arr_time','tailnum','time_hour'], axis=1)
df_clean = df_clean.dropna()


# Exportar dataset limpio
df_clean.to_csv("flights_clean.csv", index=False)
```
